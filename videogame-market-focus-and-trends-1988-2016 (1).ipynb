# 5. Assess whether the sales and scores data follow a normal distribution

It is important to assess whether the sales and scores data follow a normal distribution. A normal distribution is a common pattern observed in many real-world phenomena and is crucial for making accurate statistical inferences. To determine if the data is normally distributed, we can use statistical tests and graphical methods.

By checking for normal distribution in the sales and scores data, we can better understand the underlying patterns and variability in the data, identify potential outliers or skewness, and make informed decisions about appropriate statistical models and methods to use for analysis.

Therefore, we performe checks for normal distribution in the sales and scores data using graphical methods such as histograms and QQ plots and statistical tests such as the Shapiro-Wilk test. This allow to assess the normality of the data and make informe decisions about subsequent analysis.

##Checked sales and scores for normal distrubution (QQ plots):

```r
par(mfrow=c(2,2)) 
qqnorm(new_data$NA_sales, main = 'QQ Plot of NA Sales')
qqline(new_data$NA_sales, col=2)

qqnorm(new_data$EU_sales, main = 'QQ Plot of EU Sales')
qqline(new_data$EU_sales, col=2)

qqnorm(new_data$JP_sales, main = 'QQ Plot of JP Sales')
qqline(new_data$JP_sales, col=2)

qqnorm(new_data$global_sales, main = 'QQ Plot of Global Sales')
qqline(new_data$global_sales, col=2)
```

```r
par(mfrow=c(1,2))
qqnorm(new_data$critic_score, main = 'QQ Plot of Critic Score')
qqline(new_data$critic_score, col=2)

qqnorm(new_data$user_score, main = 'QQ Plot of User Score')
qqline(new_data$user_score, col=2)
```

QQ Plot (Quantile-Quantile Plot) is a graphical method for checking whether the distribution of data conforms to a given theoretical distribution. If the points on the QQ-plot lie approximately on the same straight line, we can conclude that the data correspond to a given (in our case, normal) distribution.

Findings where sales data follow a straight quantile line from -3 to 1.5 (in sales) and then diverge upwards may indicate outliers or outliers in the data that are not normally distributed.

It should also be noted that the QQ plot is not the only normality test method. Other methods such as the Shapiro-Wilk test and the Anderson-Darling test can be used to test for the normality of data distribution. Therefore, we will continue the statistical check and explore these additional methods to gain a better understanding of the distribution of the data. Ultimately, any decision to remove outliers should be carefully considered and based on a thorough analysis of the data and its potential impact on the overall results.

## **Checked the obtained results for longnormality (Shapiro-Wilk normality test):**

```r
shapiro.test(log(new_data$NA_sales))
shapiro.test(log(new_data$EU_sales))
shapiro.test(log(new_data$JP_sales))
shapiro.test(log(new_data$global_sales))
shapiro.test(log(new_data$critic_score))
shapiro.test(log(new_data$user_score))
```

The normality test checks whether a sample has a normal distribution or not. The test returns a test statistic and a p-value. The null hypothesis is that the data are normally distributed. A small p-value (typically less than 0.05) indicates that the null hypothesis can be rejected and the data are not normally distributed.

For the log-transformed sales data, the tests for NA_sales, EU_sales, and JP_sales return NaN values for the test statistic and p-value, which means that the data may contain NaN or Inf values. The test for global_sales returns a small p-value (0.0002423), which suggests that the data are not normally distributed.

For the log-transformed critic_score and user_score data, the tests return small p-values (< 2.2e-16) for both variables, indicating that the null hypothesis of normality can be rejected and the data are not normally distributed.

The Shapiro-Wilk normality test was performed on log-transformed sales data for North America, Europe, Japan, and global regions, as well as the critic and user scores. The p-values for North America, Europe, and Japan sales were not computed due to the presence of infinite values in the data. The global sales data showed a significant deviation from normality with a p-value of 0.0002423. The critic and user scores also showed significant deviations from normality with p-values < 2.2e-16. Therefore, these variables may not follow a normal distribution and may require non-parametric methods for further analysis.

```r
# compute skewness and kurtosis for each variable
skewness <- sapply(new_data[,c("NA_sales", "EU_sales", "JP_sales", "global_sales", "critic_score", "user_score")], skewness)
kurtosis <- sapply(new_data[,c("NA_sales", "EU_sales", "JP_sales", "global_sales", "critic_score", "user_score")], kurtosis)

# combine skewness and kurtosis into a data frame
stats_df <- data.frame(variable = c("NA_sales", "EU_sales", "JP_sales", "global_sales", "critic_score", "user_score"),
                       skewness = skewness,
                       kurtosis = kurtosis)

stats_df
```

The table shows the skewness and kurtosis values for each of the variables. Skewness is a measure of the degree of asymmetry in the distribution of the data. A value of 0 indicates a symmetrical distribution, while positive or negative values indicate a right or left-skewed distribution, respectively. Kurtosis, on the other hand, measures the degree of peakedness in the distribution of the data. A value of 3 indicates a normal distribution, while values greater than 3 indicate a more peaked distribution (leptokurtic), and values less than 3 indicate a flatter distribution (platykurtic).

Looking at the table, we can see that the sales variables (NA_sales, EU_sales, JP_sales, and global_sales) have very high skewness and kurtosis values, indicating highly skewed and peaked distributions. This suggests that these variables may not follow a normal distribution.

On the other hand, the critic_score and user_score variables have negative skewness values, indicating a left-skewed distribution, and kurtosis values close to 3, indicating a nearly normal distribution.

## **Log-normal distribution**

```r
log_NA_sales <- log(new_data$NA_sales[is.finite(log(new_data$NA_sales))])

log_NA_sales_pos <- log_NA_sales[log_NA_sales > 0]

fit <- fitdistr(log_NA_sales_pos, "lognormal")

fit$estimate
```

```r
log_EU_sales <- log(new_data$EU_sales[is.finite(log(new_data$EU_sales))])

log_EU_sales_pos <- log_EU_sales[log_EU_sales > 0]

fit <- fitdistr(log_EU_sales_pos, "lognormal")

fit$estimate
```

```r
log_JP_sales <- log(new_data$JP_sales[is.finite(log(new_data$JP_sales))])

log_JP_sales_pos <- log_JP_sales[log_JP_sales > 0]

fit <- fitdistr(log_JP_sales_pos, "lognormal")

fit$estimate
```

```r
log_global_sales <- log(new_data$global_sales[is.finite(log(new_data$global_sales))])

log_global_sales_pos <- log_global_sales[log_global_sales > 0]

fit <- fitdistr(log_global_sales_pos, "lognormal")

fit$estimate
```

The results are the meanlog and sdlog values for each variable after applying the log transformation. These values are used to describe the log-normal distribution of each variable. The meanlog represents the mean of the logarithm of the data and the sdlog represents the standard deviation of the logarithm of the data.

From these results, we can see that all the variables have negative meanlog values, indicating that the logarithm of the data is less than 1. This suggests that the original data is right-skewed. Additionally, the sdlog values for each variable indicate that there is a high degree of variability in the data.

Overall, these results suggest that the log-normal distribution is a good fit for the data, and that the data may not be normally distributed.

## **Grubbs test**

```r
grubbs.test(new_data$global_sales)
grubbs.test(new_data$NA_sales)
grubbs.test(new_data$EU_sales)
grubbs.test(new_data$JP_sales)
```

The Grubbs test is a statistical test that checks for the presence of an outlier in a univariate dataset.

The result of the test indicates that there is strong evidence to suggest that the highest value (138.6504) in the global_sales variable is an outlier. This is based on the calculated test statistic (G = 37.1284) and the p-value (< 2.2e-16), which is less than the typical significance level of 0.05.

The U-value (0.7031) is the standardized difference between the highest value and the mean of the data. A value greater than the critical value indicates that the highest value is likely an outlier.

Overall, the result suggests that the highest value in the global_sales variable is significantly different from the rest of the data and should be considered an outlier.
